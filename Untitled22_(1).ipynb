```python
dfs = [aapl_df, tsla_df, msft_df, meta_df]
for df in dfs:
    df['Date'] = pd.to_datetime(df['Date'])
```

```python
pre_covid_start = pd.to_datetime('2017-06-01')
pre_covid_end = pd.to_datetime('2019-12-31')

covid_start = pd.to_datetime('2020-01-01')
covid_end = pd.to_datetime('2020-12-31')

post_covid_start = pd.to_datetime('2021-01-01')
post_covid_end = max(df['Date'].max() for df in dfs)
```

```python
dfs_pre_covid = [df[(df['Date'] >= pre_covid_start) & (df['Date'] <= pre_covid_end)] for df in dfs]
dfs_covid = [df[(df['Date'] >= covid_start) & (df['Date'] <= covid_end)] for df in dfs]
dfs_post_covid = [df[(df['Date'] >= post_covid_start) & (df['Date'] <= post_covid_end)] for df in dfs]
```

```python
avg_stock_prices_pre_covid = [df['Close'].mean() for df in dfs_pre_covid]
avg_stock_prices_covid = [df['Close'].mean() for df in dfs_covid]
avg_stock_prices_post_covid = [df['Close'].mean() for df in dfs_post_covid]
```

```python
companies = ['Apple', 'Tesla', 'Microsoft', 'Meta Platforms']
periods = ['Pre-COVID', 'COVID', 'Post-COVID']

colors = ['blue', 'orange', 'green']

plt.bar([f"{company} - {period}" for company in companies for period in periods],
        avg_stock_prices_pre_covid + avg_stock_prices_covid + avg_stock_prices_post_covid, color=colors * len(companies))
plt.xlabel('Company - Period')
plt.ylabel('Average Stock Price')
plt.title('Average Stock Prices of Apple, Tesla, Microsoft, and Meta Platforms')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
all_data = pd.concat([aapl_df, meta_df, msft_df, tsla_df])

all_data['Date'] = pd.to_datetime(all_data['Date'])

companies = ['AAPL', 'META', 'MSFT', 'TSLA']

for company in companies:
    df_company = all_data[all_data['Company'] == company]
    df_company['Daily_Return'] = df_company['Close'].pct_change() * 100

    average_daily_return = df_company['Daily_Return'].mean()

    plt.figure(figsize=(10, 6))
    plt.plot(df_company['Date'], df_company['Daily_Return'])
    plt.axhline(average_daily_return, color='r', linestyle='dashed', label='Average Daily Return')
    plt.xlabel('Date')
    plt.ylabel('Daily Return (%)')
    plt.title(f'{company} Daily Return')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.show()

    print(f'Average daily return for {company}: {average_daily_return:.2f}%')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i in range(2):
    for j in range(2):
        company = companies[i*2 + j]
        df_company = all_data[all_data['Company'] == company]
        df_company['Daily_Return'] = df_company['Close'].pct_change() * 100

        average_daily_return = df_company['Daily_Return'].mean()

        axs[i, j].hist(df_company['Daily_Return'], bins=30, edgecolor='black')
        axs[i, j].axvline(average_daily_return, color='r', linestyle='dashed', label='Average Daily Return')
        axs[i, j].set_xlabel('Daily Return (%)')
        axs[i, j].set_ylabel('Frequency')
        axs[i, j].set_title(f'{company} Daily Return Distribution')
        axs[i, j].legend()
        axs[i, j].grid(True)

plt.tight_layout()
plt.show()
```

```python
volatility_data = []

for company in companies:
    df_company = df_covid[df_covid['Company'] == company]
    df_company['Daily_Return'] = df_company['Close'].pct_change()

    volatility = df_company['Daily_Return'].std()
    volatility_data.append((company, volatility))

volatility_data.sort(key=lambda x: x[1], reverse=True)

plt.figure(figsize=(10, 6))
plt.bar(*zip(*volatility_data))
plt.xlabel('Company')
plt.ylabel('Historical Volatility (Daily Returns)')
plt.title('Historical Volatility of Companies During COVID-19')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

print("Historical Volatility (Daily Returns) During COVID-19:")
for company, volatility in volatility_data:
    print(f"{company}: {volatility:.4f}")
```
