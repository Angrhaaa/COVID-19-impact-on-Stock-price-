```python


# Convert the 'Date' column to datetime if it's not already in that format
aapl_df['Date'] = pd.to_datetime(aapl_df['Date'])

# Filter data for the "Before COVID" period
aapl_before_covid_df = aapl_df[(aapl_df['Date'] >= '2017-06-01') & (aapl_df['Date'] <= '2019-12-31')]

# Filter data for the "During COVID" period
aapl_during_covid_df = aapl_df[(aapl_df['Date'] >= '2020-01-01') & (aapl_df['Date'] <= '2020-12-31')]

# Filter data for the "After COVID" period
aapl_after_covid_df = aapl_df[aapl_df['Date'] >= '2021-01-01']

meta_df['Date'] = pd.to_datetime(meta_df['Date'])

# Filter data for the "Before COVID" period
meta_before_covid_df = meta_df[(meta_df['Date'] >= '2017-06-01') & (meta_df['Date'] <= '2019-12-31')]

# Filter data for the "During COVID" period
meta_during_covid_df = meta_df[(meta_df['Date'] >= '2020-01-01') & (meta_df['Date'] <= '2020-12-31')]

# Filter data for the "After COVID" period
meta_after_covid_df = meta_df[meta_df['Date'] >= '2021-01-01']

msft_df['Date'] = pd.to_datetime(msft_df['Date'])

# Filter data for the "Before COVID" period
msft_before_covid_df = msft_df[(msft_df['Date'] >= '2017-06-01') & (msft_df['Date'] <= '2019-12-31')]

# Filter data for the "During COVID" period
msft_during_covid_df = msft_df[(msft_df['Date'] >= '2020-01-01') & (msft_df['Date'] <= '2020-12-31')]

# Filter data for the "After COVID" period
msft_after_covid_df = msft_df[msft_df['Date'] >= '2021-01-01']

tsla_df['Date'] = pd.to_datetime(tsla_df['Date'])

# Filter data for the "Before COVID" period
tsla_before_covid_df = tsla_df[(tsla_df['Date'] >= '2017-06-01') & (tsla_df['Date'] <= '2019-12-31')]

# Filter data for the "During COVID" period
tsla_during_covid_df = tsla_df[(tsla_df['Date'] >= '2020-01-01') & (tsla_df['Date'] <= '2020-12-31')]

# Filter data for the "After COVID" period
tsla_after_covid_df = tsla_df[tsla_df['Date'] >= '2021-01-01']


```

```python

# Calculate cumulative returns for each period
aapl_cum_returns_before = (aapl_before_covid_df['Close'] / aapl_before_covid_df['Close'].iloc[0] - 1).cumsum()
aapl_cum_returns_during = (aapl_during_covid_df['Close'] / aapl_during_covid_df['Close'].iloc[0] - 1).cumsum()
aapl_cum_returns_after = (aapl_after_covid_df['Close'] / aapl_after_covid_df['Close'].iloc[0] - 1).cumsum()

meta_cum_returns_before = (meta_before_covid_df['Close'] / meta_before_covid_df['Close'].iloc[0] - 1).cumsum()
meta_cum_returns_during = (meta_during_covid_df['Close'] / meta_during_covid_df['Close'].iloc[0] - 1).cumsum()
meta_cum_returns_after = (meta_after_covid_df['Close'] / meta_after_covid_df['Close'].iloc[0] - 1).cumsum()

msft_cum_returns_before = (msft_before_covid_df['Close'] / msft_before_covid_df['Close'].iloc[0] - 1).cumsum()
msft_cum_returns_during = (msft_during_covid_df['Close'] / msft_during_covid_df['Close'].iloc[0] - 1).cumsum()
msft_cum_returns_after = (msft_after_covid_df['Close'] / msft_after_covid_df['Close'].iloc[0] - 1).cumsum()

tsla_cum_returns_before = (tsla_before_covid_df['Close'] / tsla_before_covid_df['Close'].iloc[0] - 1).cumsum()
tsla_cum_returns_during = (tsla_during_covid_df['Close'] / tsla_during_covid_df['Close'].iloc[0] - 1).cumsum()
tsla_cum_returns_after = (tsla_after_covid_df['Close'] / tsla_after_covid_df['Close'].iloc[0] - 1).cumsum()

# Plot cumulative returns
plt.figure(figsize=(10, 6))

plt.plot(aapl_before_covid_df['Date'], aapl_cum_returns_before, label='AAPL - Before COVID')
plt.plot(aapl_during_covid_df['Date'], aapl_cum_returns_during, label='AAPL - During COVID')
plt.plot(aapl_after_covid_df['Date'], aapl_cum_returns_after, label='AAPL - After COVID')

plt.plot(meta_before_covid_df['Date'], meta_cum_returns_before, label='META - Before COVID')
plt.plot(meta_during_covid_df['Date'], meta_cum_returns_during, label='META - During COVID')
plt.plot(meta_after_covid_df['Date'], meta_cum_returns_after, label='META - After COVID')

plt.plot(msft_before_covid_df['Date'], msft_cum_returns_before, label='MSFT - Before COVID')
plt.plot(msft_during_covid_df['Date'], msft_cum_returns_during, label='MSFT - During COVID')
plt.plot(msft_after_covid_df['Date'], msft_cum_returns_after, label='MSFT - After COVID')

plt.plot(tsla_before_covid_df['Date'], tsla_cum_returns_before, label='TSLA - Before COVID')
plt.plot(tsla_during_covid_df['Date'], tsla_cum_returns_during, label='TSLA - During COVID')
plt.plot(tsla_after_covid_df['Date'], tsla_cum_returns_after, label='TSLA - After COVID')

plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title('Cumulative Returns - Before, During, and After COVID')
plt.legend()
plt.tight_layout()
plt.show()

```

```python

# Define the moving average periods
ma_days = [10, 20, 50]

# Calculate and plot moving averages for each company
for company_name, df in zip(['AAPL', 'META', 'MSFT', 'TSLA'], [aapl_df, meta_df, msft_df, tsla_df]):
    plt.figure(figsize=(12, 6))
    plt.plot(df['Date'], df['Close'], label='Close Price')

    for ma_day in ma_days:
        ma = df['Close'].rolling(window=ma_day).mean()
        plt.plot(df['Date'], ma, label=f'MA {ma_day} days')

    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(f'{company_name} Stock Prices with Moving Averages')
    plt.legend()
    plt.show()

```
